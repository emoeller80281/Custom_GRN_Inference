```python
import pandas as pd
from pathlib import Path

df = pd.read_csv("/gpfs/Labs/Uzun/SCRIPTS/PROJECTS/2024.SINGLE_CELL_GRN_INFERENCE.MOELLER/data/ground_truth_files/GSM6858447_KO_all_FRPerturb_effect_sizes.csv")
tf_df = pd.read_csv("/gpfs/Labs/Uzun/SCRIPTS/PROJECTS/2024.SINGLE_CELL_GRN_INFERENCE.MOELLER/data/databases/motif_information/hg38/TF_Information_all_motifs.txt", sep="\t")
tf_list = tf_df["TF_Name"].unique()

df_filtered = (
    df[df["Perturbed_gene"].isin(tf_list)]
    .query("Perturbed_gene != Downstream_gene")
    .query("qvalue < 0.05")
    .query("abs(Log_fold_change) > 0.25")
)

df_formatted = df_filtered[["Perturbed_gene", "Downstream_gene", "Log_fold_change"]].copy()
df_formatted = df_formatted.rename(columns={"Perturbed_gene": "source_id", "Downstream_gene": "target_id"}).drop_duplicates()

print(f"Unique TFs: {df_formatted['source_id'].nunique()}")
print(f"Unique downstream genes: {df_formatted['target_id'].nunique()}")
print(f"Number of significant perturbations: {len(df_formatted)}")

df_formatted.to_csv("/gpfs/Labs/Uzun/SCRIPTS/PROJECTS/2024.SINGLE_CELL_GRN_INFERENCE.MOELLER/data/ground_truth_files/macrophage_perturbations.csv", index=False)
```
