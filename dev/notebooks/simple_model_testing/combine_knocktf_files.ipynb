```python
from pathlib import Path
import pandas as pd

mouse_liver_knocktf_dir = Path("/gpfs/Labs/Uzun/SCRIPTS/PROJECTS/2024.SINGLE_CELL_GRN_INFERENCE.MOELLER/data/ground_truth_files/KnockTF_mouse_liver")
```

```python
df_list = []
for file in mouse_liver_knocktf_dir.glob("*.tsv"):
    df = pd.read_csv(file, sep="\t")
    print(f"File: {file.name}, Shape: {df.shape}")
    df_list.append(df)
    
total_df = pd.concat(df_list)
print(total_df.shape)
```

```python
total_df = total_df.rename(columns={"T(co)F": "source_id", "Target": "target_id"})[["source_id", "target_id"]]
total_df["source_id"] = total_df["source_id"].str.upper()
total_df["target_id"] = total_df["target_id"].str.upper()
total_df
```

```python
total_df.to_csv("/gpfs/Labs/Uzun/SCRIPTS/PROJECTS/2024.SINGLE_CELL_GRN_INFERENCE.MOELLER/data/ground_truth_files/KnockTF_mouse_liver.csv", sep=",", index=False)
```
